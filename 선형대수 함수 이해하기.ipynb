## 역행렬 구하기

      실제 역행렬의 값에 대한 부분이 단위행렬의 유사한 값이 처리되는 것을 확인할 함수인 allclose를 이용해서 처리한다.
      
      

```python
a = np.arange(1,5).reshape(2,2)
print(a)
a_i = np.linalg.inv(a)

print(a_i)
```

```python
np.dot(a,a_i)
print(np.allclose(np.dot(a, a_i), np.eye(2)))
```
