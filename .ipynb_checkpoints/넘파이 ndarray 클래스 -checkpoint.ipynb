```python
type(np.ndarray.__dict__['var'])
```

## 메소드 

```python
count = 0 
for i in dir(a) :
    
    if not i.startswith("_")  :
        
        if type(np.ndarray.__dict__[i]) == type(np.ndarray.__dict__['var']) : 
            
            count += 1
            if count % 5 == 0 :
                print()
            print(i,end=', ') 
```

### 클래스에 정의한 함수는 메소드로 처리가 된다. 정의된 함수는 메소드내에 보관

```python
class A :
    def all(iterable) :
        return all(iterable)
```

```python
i = A()
```

```python
i.all
```

```python
i.all.__func__
```

###  메소드이지만 별도의 함수 저장하는 스페셜 메소드 존재하지 않는다.

```python
a.all, a.all()
```

```python
type(a.all)
```

```python
try : 
    a.all.__func__
except Exception as e :
    print(e)
```

##  속성 

```python
count = 0 
for i in dir(a) :
    
    if not i.startswith("_")  :
        
        if type(np.ndarray.__dict__[i]) != type(np.ndarray.__dict__['var']) : 
            count += 1
            if count % 5 == 0 :
                print()
            print(i,end=', ') 
```
